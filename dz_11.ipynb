```python
def rate_movie (row):
    if row <= 2.0:
        return 'Низкий рейтинг'
    elif row <= 4.0:
        return 'Средний рейтинг'
    else:
        return 'Высокий рейтинг'
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
ratings['class'] = ratings['rating'].apply(rate_movie)
```

```python
geo = pd.read_csv('keywords.csv')
geo.head()
```

```python
geo_data = {'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']}
```

```python
def region(row):
    if ('москва' in row or 'тула' in row or 'ярославль' in row):
        return 'Центр'
    elif ('петербург' in row or 'псков' in row or 'мурманск' in row):
        return 'Северо-Запад'
    elif ('владивосток' in row or 'сахалин' in row or 'хабаровск' in row):
        return 'Дальний Восток'
    else:
        return 'undefined'
```

```python
geo['region'] = geo['keyword'].apply(lambda x: region(x))
geo
```
